# Exercise on Numpy and Matplotlib

The data in [populations.txt](http://www.scipy-lectures.org/_downloads/populations.txt) describes the populations of hares, lynxes and carrots in northern Canada during 20 years. Get the data with ```np.loadtxt``` and do the following exercises **without for-loops**.


Plot the populations for the years in the period. Add legend and axis labels to the plot.

````python
data = np.loadtxt("http://scipy-lectures.org/_downloads/populations.txt")
data = data.astype('float')
plt.figure(figsize=(15, 10))
plt.plot(data[:, 0], data[:, 1], label = "hare", c = 'g')
plt.plot(data[:, 0], data[:, 2], label = "lynx", c = 'r')
plt.plot(data[:, 0], data[:, 3], label = "carrot", c ='orange')
plt.xlabel('Years')
plt.ylabel('Population')
plt.legend()
plt.grid()
plt.show()
````

Assuming total population of hares and lynxes is 100%, plot stacked bar graph showing ratio of each speciуs for the years in the period.

````python
plt.figure(figsize=(12, 7))
plt.bar(data[:, 0], data[:, 1]/(data[:, 1] + data[:, 2] + data[:, 3]), label = "Hare")
plt.bar(data[:, 0], data[:, 2]/(data[:, 1] + data[:, 2] + data[:, 3]), label = "Lynx", color = 'r', bottom = data[:,1]/(data[:, 1] + data[:, 2] + data[:, 3])) 
plt.bar(data[:, 0], data[:, 3]/(data[:, 1] + data[:, 2] + data[:, 3]), label = "Carrot", color = "orange", bottom = (data[:,1] + data[:,2])/(data[:, 1] + data[:, 2] + data[:, 3])) 
plt.legend()
plt.show()
````

Find the mean and std of the populations of each species. Plot the historgam of population for each species and show mean values with vertical line. Arrange 3 subplots in a row.

````python
print('mean of hare population:', np.mean(data[:, 1]))
print('std of hare population:', np.std(data[:, 1]))
print('mean of lynx population:', np.mean(data[:, 2]))
print('std of lynx population:', np.std(data[:, 2]))
print('mean of carrot population:', np.mean(data[:, 3]))
print('std of carrot population:', np.std(data[:, 3]))
fig, ax = plt.subplots(1, 3, figsize = (15, 7))
ax[0].hist(data[:, 1], color = 'g', label='hare')
ax[0].axvline(data[:,1].mean(), c="black")
ax[1].hist(data[:, 2], color = 'r', label='lynx')
ax[1].axvline(data[:,2].mean(), c="black")
ax[2].hist(data[:, 3], color = 'orange', label='carrot')
ax[2].axvline(data[:,3].mean(), c="black")
plt.show()
````

Find which year each species had the largest population. Mark these years in the plot of populations.

````python
print('Hares: ' + str(int(data[np.argmax(data[:, 1])][0])))
print('Lynxes: ' + str(int(data[np.argmax(data[:, 2])][0])))
print('Carrots: ' + str(int(data[np.argmax(data[:, 3])][0])))
plt.figure(figsize=(13, 8))
plt.plot(data[:, 0], data[:, 1], label = "hare", c = 'g')
plt.plot(data[:, 0], data[:, 2], label = "lynx", c = 'r')
plt.plot(data[:, 0], data[:, 3], label = "carrot", c ='orange')
plt.scatter(data[np.argmax(data[:, 1])][0],np.max(data[:, 1]), c = 'g')
plt.scatter(data[np.argmax(data[:, 2])][0],np.max(data[:, 2]), c = 'r')
plt.scatter(data[np.argmax(data[:, 3])][0],np.max(data[:, 3]), c ='orange')
plt.xlabel('Years')
plt.ylabel('Population')
plt.legend()
plt.grid()
plt.show()
````

Find which species (hares or lynxes) has the largest population for each year. Print the result as [H, H, L, H, ...]. Plot a pie chart showing ratio of "H" and "L" values obtained.

````python
plt.figure(figsize=(11, 7))
a = np.where(data[:, 1] - data[:, 2] > 0, 'H', 'L')
p = np.array([len(np.argwhere(a == 'H')), len(np.argwhere(a == 'L'))])
plt.pie(p, labels=['hare', 'lynx'], autopct='%1.f%%')
plt.show()
print(a)
````

Find which years any of the populations is above 50000. Show time periods where populations are above 50000
in the plot of populations.

````python
plt.figure(figsize=(13, 8))
plt.xlabel('Year')
plt.ylabel('Population')
plt.plot(data[:, 0], data[:, 1], label = "hares", c = 'g')
plt.plot(data[:, 0], data[:, 2], label = "lynxes", c = 'r')
plt.plot(data[:, 0], data[:, 3], label = "carrots", c='orange')
plt.axhline(y=50000, c='black', linestyle='--', linewidth=4)
plt.scatter(data[data[:, 1] > 50000, 0], data[:, 1] [data[:, 1] > 50000], c = 'g')
plt.scatter(data[data[:, 2] > 50000, 0], data[:, 2] [data[:, 2] > 50000], c = 'r')
plt.scatter(data[data[:, 3] > 50000, 0], data[:, 3] [data[:, 3] > 50000], c='orange')
plt.legend()
plt.grid()
plt.show()
````

Find the top 2 years for each species when they had the lowest populations.

````python
data[data[:, 1:].argsort(axis=0)[:2], 0]
````

Plot the change in hare population and the number of lynxes. Find the correlation coefficient.

````python
plt.figure(figsize=(13, 8))
plt.xlabel('Year')
plt.ylabel('Number')

plt.plot(data[:, 0], np.gradient(data[:, 1]), label='Hare gradient', color='r')
plt.plot(data[:, 0], data[:, 2], label = "Lynxes", color='g')
plt.legend()
plt.grid()
plt.show()
print("correlation coefficient =", np.corrcoef(np.gradient(data[:, 1]), data[:, 2])[0, 1])
````
